***FIFA DATA***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\PYTHON\fifa_data.csv')
df
```

***HISTOGRAM***

```python
bins= [40,50,60,70,80,90,100]

plt.hist(df.Overall, bins=bins, color='orange')

plt

plt.xticks(bins)

plt.xlabel('skill level')
plt.ylabel('no of players')

plt.title('Distribution of Player Skilss in Fifa')

plt.show()
```

***PIE CHART***

```python
left=df.loc[df['Preferred Foot']== 'Left'].count().iloc[0]
left
```

```python
right=df.loc[df['Preferred Foot'] == 'Right'].count().iloc[0]
right
```

```python
labels= ['Left', 'Right']
colors= ['green', 'orange']
explode= (.1, .1)

plt.pie([left,right], labels=labels, colors=colors, autopct='%2f %%', explode=explode)

plt.title('Foot Preferrence of Fifa Players')

plt.show()
```

```python
Barcelona= df.loc[df.Club == 'FC Barcelona'] ['Overall']
Madrid= df.loc[df.Club == 'Real Madrid'] ['Overall']

labels= ['FC Barcelona', 'Real Madrid']

plt.ylabel('Fifa Overall Rating')

plt.boxplot([Barcelona, Madrid], labels=labels)

plt.show()
```
